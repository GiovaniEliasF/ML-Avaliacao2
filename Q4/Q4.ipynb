```python
# Carregar dados Iris
data = get_data('iris')

```

```python
# Inicializar PyCaret
clf = setup(data, target='species', verbose=False)

# Encontrar o melhor modelo
best_model = compare_models()

```

```python
# Finalizar e salvar o modelo
final_model = finalize_model(best_model)
save_model(final_model, 'pycaret_iris_model')
print("Modelo salvo em pycaret_iris_model.pkl")
```
